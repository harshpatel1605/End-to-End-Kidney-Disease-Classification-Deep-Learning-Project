**ensure_annotation**

```python
def get_product(x:int , y:int) -> int:
    return x*y
```

```python
get_product(x=2,y=3)
```

even if we add int in argument still if we pass string into func we will get output

```python
get_product(x=2,y='3')
```

```python
from ensure import ensure_annotations
```

```python
@ensure_annotations
def get_product(x:int , y:int) -> int:
    return x*y
```

```python
get_product(x=2,y='3')
```

```python
2*[1,2,3]
```
